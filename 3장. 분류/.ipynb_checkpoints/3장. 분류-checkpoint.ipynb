```python
# 가장 먼저, mnist 데이터를 불러올게요.
from sklearn.datasets import fetch_openml
# mnist_784의 의미는 28*28가 784이므로 그림 이미지를 가져오겠다는 말이에요.
mnist = fetch_openml('mnist_784', version=1)
# 아마 데이터는 딕셔너리로 구성되어있을테고, 데이터와 타깃값만 가져오겠습니다.
mnist.keys()
```

```python
# 아직 훈련/테스트 셋을 나누기 전이니 일단 전체 데이터로 분류할게요.
total_data = mnist['data']
total_target = mnist['target']
total_data.shape, total_target.shape
```

```python
# 자 그렇다면 데이터 중에서 첫번째 데이터를 한번 그려볼게요.
import matplotlib.pyplot as plt
# 일단 그림을 작게 만들어줄게요.
plt.figure(figsize=(2, 2))
# 기존 total_data의 형태가 데이터프레임 형태이므로, 넘파이 형태로 바꾸고, 28*28 배열로 바꾸어주도록 합시다.
plt.imshow(total_data.to_numpy()[0].reshape(-1, 28), cmap='gray_r')
plt.axis('off')
```

```python
# 5로 보이는데 한번 정답을 볼까요?
total_target[0]
```

```python
# 정답데이터를 보니, 5인건 맞는데 타입이 정수형이 아니고 문자형이네요.
# 정수형으로 바꾸어줍시다.
total_target = total_target.astype(int)
```

```python
# 7만개의 데이터 중에서 5*5 총 25개의 데이터만 그림으로 좀 그려봅시다.
fig, axes = plt.subplots(5, 5, figsize=(5, 5))
for i in range(5):
    for j in range(5):
        axes[i, j].imshow(total_data.to_numpy()[i*5+j].reshape(28, 28), cmap='gray_r')
        axes[i, j].axis('off')
```

```python
# 지금 보니 대충 적은 것 같은 글씨가 있어서 분류가 쉽지 않겠네요.
# 왼쪽 첫번째 열 맨 아래 보이시나요? 저게 4인지 뭔지 잘 모르겠네요.

# 일단 훈련데이터, 검증데이터, 테스트데이터를 나누는게 먼저입니다.
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(total_data,
                                                                      total_target,
                                                                      test_size=0.2,
                                                                      random_state=42)
train_input, val_input, train_target, val_target = train_test_split(train_input,
                                                                    train_target,
                                                                    test_size=0.2,
                                                                    random_state=42)
```

---

```python
# 가장 먼저 이진 분류를 해볼게요.
# 기준을 5로 두고, 5를 선택하는 경우와 선택하지 않는 경우로 분류하겠습니다.
# 전체 데이터 중에서 5인 것과, 그렇지 않은 것으로 분류해야되므로 타깃 데이터만 생성해주면 됩니다.
is_train_5 = (train_target==5)
is_val_5 = (val_target==5)
is_test_5 = (test_target==5)
```

```python
# 가장 먼저 확률적 경사하강법을 사용하죠.
# 기본적으로 화률적 경사하강법은 데이터가 클 때 사용하기 좋은데, 시계열 데이터도 아니므로 값의 순서에 민감하지 않습니다.
# 단, 시계열 값의 경우 데이터를 섞는 것이 좋지 않음을 기억하세요.
from sklearn.linear_model import SGDClassifier
sgdc = SGDClassifier(random_state=42)
# 그리고 나서 데이터가 5인 경우를 학습시켜볼게요.
sgdc.fit(train_input, is_train_5)
```

```python
# 훈련세트에 대한 점수가 0.97로, 5인지, 아닌지를 95% 확률로 예측한다고 하네요.
sgdc.score(train_input, is_train_5)
```

```python
# 이번에는 cross_val_score을 통해서 교차 검증한 점수를 계산해볼게요.
from sklearn.model_selection import cross_val_score
# cv가 3이므로 3번의 폴드 값에 대하여 각각의 값을 반환합니다.
cross_val_score(sgdc, train_input, is_train_5, cv=3)
```

```python
# 정확도가 96% 이상입니다. 엄청 좋은 점수네요!
# 하지만 정확도만 보면 안되고, 혼동행렬, 정밀도, 재현율 등 다양한 평가방법들을 계산해야합니다.
from sklearn.model_selection import cross_val_predict
# cross_val_predict은 교차 검증을 하면서, 예측하는 데이터를 반환합니다. 여기서는 3개의 데이터가 아니라 각 데이터의 결과를 조합하여 
# 하나의 데이터로 반환하는게 핵심입니다.
train_predict = cross_val_predict(sgdc, train_input, is_train_5, cv=3)
train_predict
```

```python
# 그래서, 원래 결과값이랑 비교를 해보자구요!
from sklearn.metrics import confusion_matrix
confusion_matrix(train_predict, is_train_5)
```

```python
# TN FP
# FN TP
# 의 구성인데요, 하나씩 설명할게요.
# TP는 Positive로 예측했는데, True인 값입니다.
# FP는 Positive로 예측했는데, False인 값입니다.
# FN은 Nagative로 예측했는데, False인 값입니다.
# TN은 Nagative로 예측했는데, True인 값입니다.
# 정확도는 accuracy는 전체 중에서 예측이 맞은 값을 나타내죠.
# 재현율은 실제값이 참인 값 중에서 참으로 예측한 값이고.
# 정밀도는 예측한 값이 참인 값 중에서 실제 참인 값입니다.
# 따라서 정확도만 높다고 좋은 모델이 아니라, 정밀도는 물론 재현율도 좋아야하는거죠.

# 그래서 위의 데이터를 보면 5를 예측하고 맞은 경우가 2660건
# 5를 예측하고 틀린 값이 1341건
# 5가 아니라고 예측했는데 틀린값이 234건
# 5가 아니라고 예측했는데 맞은값이 40565건이 되는거죠.
```

```python
# 그렇다면 정밀도와 재현율을 구현해볼까요?
from sklearn.metrics import precision_score, recall_score, accuracy_score
print('정밀도 : {}'.format(precision_score(is_train_5, train_predict)))
print('재현율 : {}'.format(recall_score(is_train_5, train_predict)))
print('정확도 : {}'.format(accuracy_score(is_train_5, train_predict)))
```

```python
# 내가 참으로 예측한 값이 참인 확률은 높지만, 실제 참인 값 중에서 참으로 예측한 값이 낮다는 것을 알 수 있습니다.
# 이 값을 잘 조절해야하는거죠.
# 이렇게 두 값을 비교하는 것 보다, f1 score로 한번에 관리하는게 더 쉽습니다.
from sklearn.metrics import f1_score
print('f1 점수 : {}'.format(f1_score(is_train_5, train_predict)))
```

```python
# 하지만 재현율과 정밀도 둘 다 높은 값을 유지할 수 없습니다. 
# 이것을 바로 정밀도/재현율 트레이드오프라고 이야기합니다. 하나의 값이 높다면, 하나의 값은 낮아야하는거죠.
# SGDClassifier(확률적 경사하강법)은 결정 함수를 사용하여 각 샘플의 점수를 계산합니다.
# 이 점수가 임계값보다 크면 샘플을 양성으로 할당하고, 그렇지 않으면 음성이라고 할당하는거죠.
# 이 임계값의 이동에 따라 재현율이 커지기도, 정밀도가 커지기도 하는거죠.
train_scores = sgdc.decision_function([train_input.to_numpy()[2]])
train_scores, train_target[2]
```

```python
# 이렇게 임계값에 넘지 못하면 False를 반환하는거죠
train_data_predict = (train_scores > 0)
train_data_predict
```

```python
# 임계값을 낮추면, True를 반환하겠죠?
train_data_predict = (train_scores < 10000)
train_data_predict
```

```python
# 그렇다면, 어떻게 임계값을 적절하게 정할 수 있을까요?
# 이를 위해서는 cross_val_predict 함수를 사용해 모든 샘플들의 점수를 구해야합니다.
train_scores = cross_val_predict(sgdc, train_input, is_train_5, cv=3,
                                 method='decision_function')
train_scores
```

```python
# 위에서 사용하는 decision_function이 더 신뢰성이 높습니다.
sgdc.decision_function(train_input)
```

```python
# 이제 precision_recall_curve를 통해 precision과 recalls, threshold를 찾을 수 있습니다.
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(is_train_5, train_scores)
```

```python
# 처음 값을 제외하는 이유는 임계값은 재현율과 정밀도의 개수보다 하나 적습니다.
# 왜냐하면 임계값은 초기값을 포함하지 않는데, 재현율과 정밀도는 임계값 이전의 초기값을 먼저 계산하기 때문입니다.
plt.plot(thresholds, precisions[1:])
plt.plot(thresholds, recalls[1:])
plt.legend(['precisions', 'recalls'])
```

```python
# 만약, 재현율이 90%를 달성하는 것이 목표라면 임계값이 대략 -12500정도가 되어야겠죠?
import numpy as np
thresholds_90_recall = thresholds[np.argmin(recalls > 0.9)]
thresholds_90_recall
```

```python
train_predict_90 = (train_scores > thresholds_90_recall)
```

```python
precision_score(is_train_5, train_predict_90)
```

```python
recall_score(is_train_5, train_predict_90)
```

```python
# 이번에는 ROC curve입니다.
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(is_train_5, train_scores)
```

```python
plt.plot(fpr, tpr)
plt.plot([0, 1])
```

```python
# 아래부분의 점수를 보려면 roc_auc_score를 불러오면 됩니다.
from sklearn.metrics import roc_auc_score
roc_auc_score(is_train_5, train_scores)
```

```python
# 이번에는 랜덤포레스트로 훈련을 시켜볼게요.
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
train_scores = cross_val_predict(rfc, train_input, is_train_5, cv=3,
                                 method='predict_proba')
```

```python
# 양성의 값만을 사용할게요.
train_scores = train_scores[:, 1]
fpr_random, tpr_random, thresholds_random = roc_curve(is_train_5, train_scores)
```

```python
plt.plot(fpr, tpr)
plt.plot(fpr_random, tpr_random)
plt.plot([0, 1])
plt.legend(['SGD', 'Random', '0~1'])
```
